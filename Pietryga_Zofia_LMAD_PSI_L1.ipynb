```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Zad. 1**

```python
df_iris = pd.read_csv("iris.csv")
df_iris.head()
```

```python
df_iris.describe()
```

```python
print(df_iris.isnull().any())
```

```python
iris_normalized_0_1 = df_iris.copy()

for column in df_iris.columns[:-1]:
    X_min = df_iris[column].min()
    X_max = df_iris[column].max()
    iris_normalized_0_1[column] = (df_iris[column] - X_min) / (X_max - X_min)

print(iris_normalized_0_1.head())
```

```python
iris_normalized_1_1 = df_iris.copy()

for column in df_iris.columns[:-1]:
    X_min = df_iris[column].min()
    X_max = df_iris[column].max()
    iris_normalized_1_1[column] = 2 * ((df_iris[column] - X_min) / (X_max - X_min)) - 1

print(iris_normalized_1_1.head())
```

```python
iris_standardized = df_iris.copy()

for column in df_iris.columns[:-1]:
    mean = df_iris[column].mean()
    std = df_iris[column].std()
    iris_standardized[column] = (df_iris[column] - mean) / std

print(iris_standardized.head())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df_iris, ax=axes[0, 0])
axes[0, 0].set_title('Dane pierwotne')
axes[0, 0].set_xlabel('petal_length [cm]')
axes[0, 0].set_ylabel('petal_width [cm]')

sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_normalized_0_1, ax=axes[0, 1])
axes[0, 1].set_title('Dane znormalizowane [0, 1]')
axes[0, 0].set_xlabel('petal_length [cm]')
axes[0, 0].set_ylabel('petal_width [cm]')

sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_normalized_1_1, ax=axes[1, 0])
axes[1, 0].set_title('Dane znormalizowane [-1, 1]')
axes[0, 0].set_xlabel('petal_length [cm]')
axes[0, 0].set_ylabel('petal_width [cm]')

sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_standardized, ax=axes[1, 1])
axes[1, 1].set_title('Dane standaryzowane')
axes[0, 0].set_xlabel('petal_length [cm]')
axes[0, 0].set_ylabel('petal_width [cm]')

plt.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df_iris, ax=axes[0, 0])
axes[0, 0].set_title('Dane pierwotne')
axes[0, 0].set_xlabel('sepal_length [cm]')
axes[0, 0].set_ylabel('sepal_width [cm]')

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_normalized_0_1, ax=axes[0, 1])
axes[0, 1].set_title('Dane znormalizowane [0, 1]')
axes[0, 0].set_xlabel('sepal_length [cm]')
axes[0, 0].set_ylabel('sepal_width [cm]')

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_normalized_1_1, ax=axes[1, 0])
axes[1, 0].set_title('Dane znormalizowane [-1, 1]')
axes[0, 0].set_xlabel('sepal_length [cm]')
axes[0, 0].set_ylabel('sepal_width [cm]')

sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_standardized, ax=axes[1, 1])
axes[1, 1].set_title('Dane standaryzowane')
axes[0, 0].set_xlabel('sepal_length [cm]')
axes[0, 0].set_ylabel('sepal_width [cm]')

plt.tight_layout()

plt.show()
```

**Zad. 2**

```python
df_zad2 = pd.read_csv("Zad2_L1.csv", sep=';')
df_zad2.head()
```

```python
df_zad2.columns = ['Wavenumber [cm^-1]'] + [f"t{i}" for i in range(df_zad2.shape[1] - 1)]
df_zad2 = df_zad2.drop(0)
df_zad2 = df_zad2.applymap(lambda x: x.replace(",", ".") if isinstance(x, str) else x)
df_zad2 = df_zad2.astype(float)
```

```python
# Indeks wiersza odpowiadajcy numerowi fali 985 cm^-1
target_wavenumber = 985
index_985 = (df_zad2['Wavenumber [cm^-1]'] - target_wavenumber).abs().idxmin()
print(index_985)

# Normalizacja kadego widma wzgldem wartoci przy 985 cm^-1
df_zad2_normalized = df_zad2.copy()
for col in df_zad2_normalized.columns[1:]:
    df_zad2_normalized[col] = df_zad2_normalized[col] / df_zad2_normalized.loc[index_985, col]

# Wywietl dane po normalizacji
print(df_zad2_normalized.head())
```

```python
# Surowe widma
plt.figure(figsize=(10, 6))
for col in df_zad2.columns[1:]:
    plt.plot(df_zad2['Wavenumber [cm^-1]'], df_zad2[col], label=f"Raw {col}")

plt.title("Pierwotne wartoci widm Ramana")
plt.xlabel("Wavenumber [cm^-1]")
plt.ylabel("Intensity [a.u.]")
plt.legend()
plt.show()

# Znormalizowane widma
plt.figure(figsize=(10, 6))
for col in df_zad2_normalized.columns[1:]:
    plt.plot(df_zad2_normalized['Wavenumber [cm^-1]'], df_zad2_normalized[col], label=f"Normalized {col}")

plt.title("NPomiary widm Ramana znormalozowane do wartoci przy widmie 985[cm^-1]")
plt.xlabel("Wavenumber [cm^-1]")
plt.ylabel("Normalized Intensity [a.u.]")
plt.grid(True)
plt.legend()
plt.show()
```

**Zad. 3**

```python
df_zad3 = pd.read_csv("Zad3_L1.csv", sep=';')

df_zad3 = df_zad3.applymap(lambda x: x.replace(",", ".") if isinstance(x, str) else x)
df_zad3 = df_zad3.astype(float)

print(df_zad3.head())
```

```python
# Normalizacja widm wzgldem pola powierzchni
df_zad3_normalized = df_zad3.copy()
for col in df_zad3_normalized.columns[1:]:
    integral = df_zad3[col].sum()
    df_zad3_normalized[col] = df_zad3[col] / integral

print(df_zad3_normalized.head())
```

```python
# Surowe widma
plt.figure(figsize=(12, 6))
for col in df_zad3.columns[1:]:
    plt.plot(df_zad3['Wavenumber [cm^-1]'], df_zad3[col], label=f"Raw {col}")

plt.title("Raw FTIR Spectra")
plt.xlabel("Wavenumber [cm^-1]")
plt.ylabel("Absorbance [a.u.]")
plt.legend()
plt.show()

# Znormalizowane widma
plt.figure(figsize=(12, 6))
for col in df_zad3_normalized.columns[1:]:
    plt.plot(df_zad3_normalized['Wavenumber [cm^-1]'], df_zad3_normalized[col], label=f"Normalized {col}")

plt.title("Normalized FTIR Spectra")
plt.xlabel("Wavenumber [cm^-1]")
plt.ylabel("Normalized Absorbance [a.u.]")
plt.legend()
plt.show()
```
